## vraag 1
Beeldherekenning bij de mens bestaat uit verschillende onderdelen. Bij beeldanalyse worden vorm, orientatie, textuur en bewegingsrichting geanalyseerd. Op basis van deze anayse word een object in een categorie ingegeeld en herkent.
Hierbij wordt voor ieder los object geanalyseerd wat dit is.
Object herkenning vind pas plaats wanneer onze aandacht op een object is gericht. Dan worden alle met de herkenning van dat object bezig zijnde cellen actiever. [1]
Een groot deel van onze hersennen wordt gebruikt voor het herkennen van beeld. Daarmee blijkt eens te meer hoe belangrijk dit is. Het doel van het visuele systeem is een complex beeld op te delen in verschillende makkelijker te begrijpen onderdelen. Deze kunnen vervolgens makkelijker herkent worden op basis van wegingsfactoren.[2]

## vraag 2
De computer heeft geen leven vol ervaring met het zien en herkennen van vorm. Ook heeft een computer, afhankelijk van de wijze waarop informatie gepresenteerd wordt, minder context. 

## vraag 3
Het probleem met het alleen gebruiken van pixels het het invariance problem. Als een object verandert van posititie of schaal is dit binnen pixels een compleet nieuw beeld, De computer kan een beeld van een 1 niet recuderen tot een min of meer verticale lijn, zoals wij dat kunnen.
We zullen het met pixels moeten doel, maar voor betrouwbare output moeten tussenstappen worden gemaakt.
