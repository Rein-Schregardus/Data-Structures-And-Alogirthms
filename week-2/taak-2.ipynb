Laad de mnist data uit **tensorflow.keras.datasets**. 

```python
from tensorflow.keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

import matplotlib
import matplotlib as mpl

print("Loading set")
# Load MNIST dataset
(train_images, train_labels), (_, _) = mnist.load_data()
train_images = train_images.astype('float32') / 255.0
print("Loaded")
```

## Taak 2: Batchstatistieken
- Bereken de **gemiddelde waarde en standaardafwijking** van de pixelwaarden voor alle cijfers in de trainingsset (dus van 0 t/m 9).
- Print het totale gemiddelde/standaardafwijking en het gemiddelde/standaardafwijking per cijfer.
- Beschrijf wat je hier ziet? 
- Wat zou je kunnen met de stdev en de gemiddelde waarden?


```python
# Methods
def ReportOnNumber(number: int = -1):
    used_images = list()
    if number == -1:
        used_images = train_images
    else:
        used_images = [x for i, x in enumerate(train_images) if train_labels[i] == number] # only for specified number

    CreateAverageImage(used_images)

def CreateAverageImage(images: list):

    composite = np.mean(images, axis=0)
    std = np.std(images, axis=0)

    axes = plt.subplot()
    plt.imshow(composite, cmap="gray")

    height, width = composite.shape

    for y in range(height):
        for x in range(width):
            axes.text(x, y, f"{std[y,x]:.2f}",
                       ha="center", va="center", color="r", fontsize=6)

    fig = matplotlib.pyplot.gcf()
    fig.set_size_inches(10, 10)

    plt.show()

    plt.imshow(std, cmap="gray")
    plt.show()

ReportOnNumber(1)
```

```python
ReportOnNumber(-1)

for n in range(0, 10):
    ReportOnNumber(n)
```
