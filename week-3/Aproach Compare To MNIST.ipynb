### Deel 4: Testen

Je laat de boom op een aantal nieuwe afbeeldingen een voorspelling doen.

Bijvoorbeeld:

- toon 10 nieuwe cijfers
- laat de boom voorspellen
- vergelijk met het echte label
- bespreek wat opvalt

```python
count = 0
mistakes = 0;
mistake_certainty = 0
correct_certainty = 0
for i, _ in enumerate(images):
    count += 1
    if count > 100:
        break;
    differances = calculate_min_differances(images[i])
    mean = np.mean(list(differances.values()))
    minimum = min(differances.values())
    certainty = mean / minimum
    guess = min(differances, key=differances.get)
    print(f"label: {labels[i]} guess: {guess}")
    print(mean)
    print(minimum)
    print(certainty)
    if (guess != labels[i]):
        mistakes += 1
        mistake_certainty += certainty
    else:
        correct_certainty += certainty
print("Finished!")
print(f"mistakes: {mistakes}")
print(f"mistake certainty {mistake_certainty / mistakes}")
print(f"correct certainty {correct_certainty/ (count - mistakes)}")

```
